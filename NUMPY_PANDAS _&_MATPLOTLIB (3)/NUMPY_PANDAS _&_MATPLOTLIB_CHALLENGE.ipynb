```python
d5 = data[(data.body=='SUV')&(data.saledate=='Fri Dec 19 2014 09:00:00 GMT-0800 (PST)')]
d5
```

```python
pvt5 = pd.pivot_table(d5,index=['body'],values=['sellingprice'],aggfunc='sum')
pvt5
```

```python
d51 = data[(data.body=='SUV')&(data.saledate=='Thu Jan 15 2015 04:30:00 GMT-0800 (PST)')]
d51
```

```python
pvt51 = pd.pivot_table(d51,index=['body'],values=['sellingprice'],aggfunc='sum')
pvt51
```

```python
print(pvt5)
print(pvt51)
pvtd = pvt5-pvt51
print('diference between the given period : \n',pvtd)
```

```python
c51 = pvtd/pvt5*100
c51
```

```python
c52 = pvtd/pvt51*100
c52
```

```python
percentchange = c52-c51
percentchange
print('therefore, the percentage sales difference of SUV between the given period is : \n',percentchange)
```
