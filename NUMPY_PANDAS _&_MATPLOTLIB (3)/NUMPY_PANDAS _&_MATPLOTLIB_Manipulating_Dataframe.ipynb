# Task-1)

## Assume that you have some data of customers with Cust_ID in one sheet of an excel file and some more data with Cust_ID in a second sheet of the same excel file. How can you join these data?
### Read data from first sheet of excel file 'customer.xlsx'

```python
#Reading data from excel file
import pandas as pd
df_cust1 = pd.read_excel('customer.xlsx', sheet_name=0)
df_cust1
```

### Read the data from second sheet of excel file.

```python
#Reading data from excel file
df_cust2 = pd.read_excel('customer.xlsx', sheet_name=1)
df_cust2         
```

## 1. Inner Join

```python
#inner join - use intersection of index observed in both dataframes
df_cust1.set_index('Cust_ID').join(df_cust2.set_index('Cust_ID'), on='Cust_ID', how='inner')
```

## 2. Outer Join

```python
#outer join - use union of index observed in both dataframes
df_cust1.set_index('Cust_ID').join(df_cust2.set_index('Cust_ID'), on='Cust_ID', how='outer')
```

## 3. Right Join

```python
#right join - use only the index found in right dataframes
df_cust1.set_index('Cust_ID').join(df_cust2.set_index('Cust_ID'), on='Cust_ID', how='right')
```

## 4. Left Join

```python
#left join - use only the index found in left dataframes
df_cust1.set_index('Cust_ID').join(df_cust2.set_index('Cust_ID'), on='Cust_ID', how='left')
```

# Task-2)

## Assume that you have some data of customers with Cust_ID in one sheet of an excel file and some more data with Cust_ID in a second sheet of same excel file. How can you join these data?
### Read data from first sheet of excel file 'Ecommerce_data.xlsx'

```python
#Reading data from excel file
df_cust = pd.read_excel('Ecommerce_data.xlsx', sheet_name=0)
df_cust
```

### Read data from second sheet of excel file

```python
df_order = pd.read_excel('Ecommerce_data.xlsx', sheet_name=1)
df_order
```

## 1. Inner Merge

```python
#Merging dataframe on common customer_ID
pd.merge(df_cust, df_order, on='Cust_ID')
#merges the common data only
```

## 2. Outer Merge

```python
#Outer merge
pd.merge(df_cust, df_order, on='Cust_ID', how='outer')
#merges every thing frm both the data.

# If we different column names in two data frames and if we want to merge....
# use left_on and right_on parameters of merge function
```

## 3. Right Merge

```python
#Right merge
pd.merge(df_cust, df_order, on='Cust_ID', how='right')
```

## 4. Right Merge

```python
#Left merge
pd.merge(df_cust, df_order, on='Cust_ID', how='left')
```
