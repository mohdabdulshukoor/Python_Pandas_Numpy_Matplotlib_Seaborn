```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
cars = pd.read_excel('cars_ds.xlsx')
cars
```

```python
pd.set_option('display.max_rows', None)
```

```python
cars.keys()
```

# TASK 1

```python
t1a = pd.pivot_table(cars,index=['Make'],values=['City_Mileage_km_litre'])
print(t1a)

t1b = pd.pivot_table(cars,index=['Make'],values=['Highway_Mileage_km_litre'])
print(t1b)
```

```python
ta = t1a.sort_values(['City_Mileage_km_litre'],ascending=False)[:10]
print(ta)

tb = t1b.sort_values(['Highway_Mileage_km_litre'],ascending=False)[:10]
print(tb)

#Displaying the cars (with high mileage per km)
```

# TASK 2

```python
tx = t1a.sort_values(['City_Mileage_km_litre'],ascending=True)[:10]
print(tx)

ty = t1b.sort_values(['Highway_Mileage_km_litre'],ascending=True)[:10]
print(ty)

#Displaying the other categories cars (with low mileage per km)
```
